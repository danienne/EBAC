```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head(20).to_string())
```

```python
#Gráfico de barra
plt.figure(figsize=(10,6))
df['Nota'].value_counts().plot(kind='bar', color='#6A5ACD')
plt.title('Notas para os produtos')
plt.xlabel('Nota')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.show()
```

```python
#Gráfico de pizza
generos_desejados = ['Feminino', 'Masculino', 'Bebês', 'Meninas', 'Meninos']

df['Genero_Ajustado'] = df['Gênero'].apply(
    lambda x: x if x in generos_desejados else 'Outros'
)

counts = df['Genero_Ajustado'].value_counts()

x= counts.index
y= counts.values

plt.figure(figsize=(10,6))
plt.pie(y,labels=x, autopct='%.1f%%', startangle=190)
plt.title('Porcentagem de vendas por gênero')
plt.show()
```

```python
#Gráfico de dispersão
plt.hexbin(df['Preço'],df['Nota'], gridsize=40, cmap='Blues')
plt.colorbar(label='Contagem dentro do bin')
plt.xlabel('Preço')
plt.ylabel('Nota')
plt.title('Dispersão de Marca e Preço')
plt.show()
```

```python
#Gráfico de densidade
plt.figure(figsize=(10,6))
sns.kdeplot(df['Nota'], fill=True, color='#8B008B')
plt.title('Densidade de Notas')
plt.xlabel('Nota')
plt.show()
```

```python
#Gráfico de Histograma
sns.pairplot(df[['Preço', 'Nota','N_Avaliações', 'Desconto']])
plt.title('Gráfico Histograma')
plt.show()
```

```python
#Gráfico de Regressão
sns.regplot(x='Preço', y='Nota', data=df, color='#DAA520', scatter_kws={'alpha':0.5, 'color':'#B8860B'})
plt.title('Regressão de Preço por Nota')
plt.xlabel('Preço')
plt.ylabel('Nota')
plt.show()
```

```python
#Mapa de calor
df_corr = df[['Preço', 'Nota','N_Avaliações', 'Desconto']].corr()

sns.heatmap(df_corr, annot=True, fmt=".2f")
plt.title('Mapa de Calor da Correlação entre Variáveis')
plt.show()
```
